**HERE IS THE CODE FROM WHERE THESE ANALYSIS COMES FROM**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Enrollments_Dataset - Enrollments by Parents.csv')
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
columns_to_drop = ['Unnamed: 35', 'Unnamed: 36', 'Unnamed: 37']
data_cleaned = data.drop(columns=columns_to_drop, errors='ignore')
```

```python
missing_threshold = 0.5
missing_perc = data_cleaned.isnull().mean()
data_cleaned = data_cleaned.loc[:, missing_perc < missing_threshold]
```

```python
categorical_cols = data_cleaned.select_dtypes(include='object').columns
numerical_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
```

```python
data_cleaned[categorical_cols] = data_cleaned[categorical_cols].fillna('Unknown')
data_cleaned[numerical_cols] = data_cleaned[numerical_cols].fillna(0)
```

```python
data_cleaned = data_cleaned.drop_duplicates()
```

```python
data_cleaned.shape
```

```python
city_counts = data_cleaned['City'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
city_counts[:10].plot(kind='bar', color='skyblue')
plt.title('Top 10 City by Enrollment')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
data_cleaned = data_cleaned[data_cleaned['Attended?'] != 'Duplicate entry']
```

```python
attendance_rate = data_cleaned['Attended?'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(6, 4))
attendance_rate.plot(kind='bar', color='orange')
plt.title('Attendance Rate')
plt.xlabel('Attendance')
plt.ylabel('Percentage')
plt.show()
```

```python
numeric_scores = ['Score - Concept', 'Score - Interesting', 'Score - Another class', 'Final score']
plt.figure(figsize=(8, 6))
sns.heatmap(data_cleaned[numeric_scores].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Scores')
plt.show()
```

```python
class_type_counts = data_cleaned['Type'].value_counts()
plt.figure(figsize=(6, 4))
class_type_counts.plot(kind='bar', color='green')
plt.title('Class Types Distribution')
plt.xlabel('Class Type')
plt.ylabel('Count')
plt.show()
```

```python
source_counts = data_cleaned['Source'].value_counts()
plt.figure(figsize=(10, 6))
source_counts[:10].plot(kind='bar', color='purple')
plt.title('Top 10 Sources of Enrollments')
plt.xlabel('Source')
plt.ylabel('Count')
plt.show()
```

```python
timezone_counts = data_cleaned['Timezone'].value_counts()
plt.figure(figsize=(10, 6))
timezone_counts[:10].plot(kind='bar', color='teal')
plt.title('Top 10 Timezones by Enrollment')
plt.xlabel('Timezone')
plt.ylabel('Count')
plt.show()
```

```python
if 'Start time (in PST)' in data_cleaned.columns:
    data_cleaned['Start Hour'] = pd.to_datetime(data_cleaned['Start time (in PST)'], errors='coerce').dt.hour
    start_hour_counts = data_cleaned['Start Hour'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    start_hour_counts.plot(kind='bar', color='coral')
    plt.title('Class Start Times Distribution')
    plt.xlabel('Hour of Day (PST)')
    plt.ylabel('Count')
    plt.show()
```

```python
if 'Day' in data_cleaned.columns:
    day_counts = data_cleaned['Day'].value_counts()
    plt.figure(figsize=(8, 5))
    day_counts.plot(kind='bar', color='gold')
    plt.title('Popular Days for Classes')
    plt.xlabel('Day')
    plt.ylabel('Count')
    plt.show()
```

```python
if 'Timezone' in data_cleaned.columns and 'Start Hour' in data_cleaned.columns:
    timezone_hour = data_cleaned.groupby(['Timezone', 'Start Hour']).size().unstack(fill_value=0)
    plt.figure(figsize=(12, 8))
    sns.heatmap(timezone_hour, cmap='YlGnBu', annot=False)
    plt.title('Heatmap of Timezones vs. Start Hours')
    plt.xlabel('Start Hour (PST)')
    plt.ylabel('Timezone')
    plt.show()
```

```python
if 'Teacher Name' in data_cleaned.columns and 'Final score' in data_cleaned.columns:
    teacher_scores = data_cleaned.groupby('Teacher Name')['Final score'].mean().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    teacher_scores[:10].plot(kind='bar', color='navy')
    plt.title('Top 10 Teachers by Average Final Score')
    plt.xlabel('Teacher')
    plt.ylabel('Average Final Score')
    plt.show()
```

```python
if 'Type' in data_cleaned.columns and 'Final score' in data_cleaned.columns:
    class_scores = data_cleaned.groupby('Type')['Final score'].mean().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    class_scores[:10].plot(kind='bar', color='darkgreen')
    plt.title('Top 10 Class Types by Average Final Score')
    plt.xlabel('Class Type')
    plt.ylabel('Average Final Score')
    plt.show()
```

```python
if 'Teacher Name' in data_cleaned.columns and 'Type' in data_cleaned.columns:
    teacher_class_performance = data_cleaned.groupby(['Teacher Name', 'Type'])['Final score'].mean().reset_index()
    top_performances = teacher_class_performance.sort_values(by='Final score', ascending=False).head(10)
    print("Top Teacher-Class Combinations by Average Final Score:\n", top_performances)
```
